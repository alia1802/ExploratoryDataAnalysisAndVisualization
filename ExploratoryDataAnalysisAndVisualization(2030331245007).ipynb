```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('train.csv')
```

# 1. countplot/ Barchart

```python
sns.countplot(x='Survived', data=df)
```

```python
sns.countplot(x='Survived', data=df,hue='Sex')
```

# 2 line chart basically used for numeric type of variable

```python
sns.lineplot(x='Age',y='Fare',data=df,color="maroon")
```

# 3. scatter plot

```python
df.plot.scatter('Age','Fare',color="black")
```

```python
df.plot.scatter('Age','Survived',color="red")
```

# 4 Histogram

```python
df['Age'].hist(bins=8)
```

# 5 Pie chart

```python
import matplotlib.pyplot as plt
sizes=df['Survived'].value_counts()
fig1,ax1=plt.subplots()
ax1.pie(sizes,labels=['Not Survived','Survived'],autopct='%1.1f%%',shadow=True)
plt.show()
```
